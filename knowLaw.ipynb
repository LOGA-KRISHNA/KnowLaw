```python
# Import necessary libraries
from transformers import AutoModelForCausalLM, AutoTokenizer, BitsAndBytesConfig
import torch
import gradio as gr

# Replace with your actual model path
model_name = "logakrishna/KnowLaw"

# Load tokenizer and model with quantization configuration
tokenizer = AutoTokenizer.from_pretrained(model_name)

# Quantization configuration for loading the model in 4-bit
quantization_config = BitsAndBytesConfig(
    load_in_4bit=True,  # Enable 4-bit quantization
    bnb_4bit_compute_dtype=torch.bfloat16,  # Set compute dtype to bfloat16
    bnb_4bit_quant_type="nf4",  # Set quantization type to nf4
)

model = AutoModelForCausalLM.from_pretrained(
    model_name, quantization_config=quantization_config
)

# Set device (use GPU if available)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)

# Remove the indentation from this function
def generate(input_text):
    # Tokenize and generate response
    inputs = tokenizer(input_text, return_tensors="pt").to(device)
    output = model.generate(**inputs, max_length=200)
    response = tokenizer.decode(output[0], skip_special_tokens=True)
    return response

# Correct the function name in gr.Interface
iface = gr.Interface(fn=generate, inputs="text", outputs="text")
iface.launch(share=True)  # `share=True` gives you a public link
```
